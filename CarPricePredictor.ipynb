# MODEL


```python
X=car.drop("Price",axis=1)
```

```python
y=car[["Price"]]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
X_encoded=ohe.transform(X[['name','company','fuel_type']])
#.to array() to convert sparse onehot encoded matrix x_encoded to a dense numpy array
X_encoded_df=pd.DataFrame(X_encoded.toarray(),columns=ohe.get_feature_names_out())
```

```python
X_new=pd.concat([X.drop(['name','company','fuel_type'],axis=1),X_encoded_df],axis=1)
```

```python
X_new
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_new,y,test_size=0.2)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
predictions=model.predict(X_test)
rows=range(len(predictions))
columns=['Price']
output=pd.DataFrame(predictions,rows,columns)
```

```python
output
```

```python
r2_score(predictions,y_test)
```

```python
scores=[]
for i in range(1000):
    from sklearn.model_selection import train_test_split
    X_train,X_test,y_train,y_test=train_test_split(X_new,y,test_size=0.2,random_state=i)
    model=LinearRegression()
    model.fit(X_train,y_train)
    predictions=model.predict(X_test)
    scores.append(r2_score(predictions,y_test))
```

```python
val=max(scores)
```
